```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/jobs")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
        """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}

    ### INSTRUCTION:
    The scraped text is from the careers page of a website.

    Extract job postings and return a JSON list where each object contains:
    - role (string)
    - experience (string or null)
    - skills (list of strings — MUST NOT be null. If skills aren't explicitly present,
      infer 5–7 relevant skills based on the job role.)
    - description (string)

    Ensure:
    - skills is ALWAYS a list of valid, meaningful skills
    - Only return VALID JSON (NO MARKDOWN, NO TEXT, NO EXTRA CHARACTERS)
    """
)

chain_extract = prompt_extract | llm 
res = chain_extract.invoke(input={'page_data':page_data})
type(res.content)
```
